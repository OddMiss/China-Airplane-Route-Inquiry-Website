```python
# 先分别获取地址的经纬度
import requests
import json
import pandas as pd

# 调用函数进行地址经纬度信息的获取
Airplane_all = pd.read_excel("国内航班数据new.xls")
# df["position"] = df["name"].apply(lambda x: gis_post(x))
```

```python
Airplane_route = Airplane_all[
    [
        "departure_city",
        "departure_cy",
        "departure_cx",
        "landing_city",
        "landing_cy",
        "landing_cx",
    ]
]
Airplane_route = Airplane_route.drop_duplicates(
    subset=[
        "departure_city",
        "departure_cy",
        "departure_cx",
        "landing_city",
        "landing_cy",
        "landing_cx",
    ]
)
Airplane_depature_city = Airplane_route[["departure_city", "departure_cy", "departure_cx"]]
Airplane_depature_city = Airplane_depature_city.rename(columns={'departure_city': 'city',
                                                                'departure_cy': 'cy',
                                                                'departure_cx': 'cx'})
Airplane_depature_city = Airplane_depature_city.drop_duplicates(subset=['city'])

Airplane_landing_city = Airplane_route[["landing_city", "landing_cy", "landing_cx"]]
Airplane_landing_city = Airplane_depature_city.rename(columns={'landing_city': 'city',
                                                                'landing_cy': 'cy',
                                                                'landing_cx': 'cx'})
Airplane_depature_city = Airplane_landing_city.drop_duplicates(subset=['city'])
Airplane_city = pd.concat([Airplane_depature_city, Airplane_landing_city], axis=0)
Airplane_city = Airplane_city.drop_duplicates(subset=['city'])
Airplane_city = Airplane_city.reset_index(drop=True)
```

```python
Airplane_city['city'].to_csv('Airplane_city.csv', index=False)
```

```python
Airplane_route.drop(columns=["departure_cy", "departure_cx", "landing_cy", "landing_cx"], inplace=True)
Airplane_route.reset_index(drop=True, inplace=True)
```

```python
Airplane_route.to_csv('Airplane_route.csv', index=False)
```
